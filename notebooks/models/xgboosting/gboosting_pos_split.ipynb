```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import mean_absolute_error as mae, mean_absolute_percentage_error as mape, root_mean_squared_error as rmse, r2_score
```

```python
X_train = pd.read_csv('csv/final_datasets/for_xgb/train_test_val/predictors_train.csv')
X_test = pd.read_csv('csv/final_datasets/for_xgb/train_test_val/predictors_test.csv')
X_val = pd.read_csv('csv/final_datasets/for_xgb/train_test_val/predictors_val.csv')
for col in ('Nation', 'Pos', 'Squad', 'league'):
    X_train[col] = X_train[col].astype('category')
    X_test[col] = X_test[col].astype('category')
    X_val[col] = X_val[col].astype('category')

players_train = pd.read_csv('csv/final_datasets/for_xgb/train_test_val/train_players_names.csv')
players_test = pd.read_csv('csv/final_datasets/for_xgb/train_test_val/test_players_names.csv')
players_val = pd.read_csv('csv/final_datasets/for_xgb/train_test_val/val_players_names.csv')

y_train = pd.read_csv('csv/final_datasets/for_xgb/train_test_val/target_train.csv')
y_test = pd.read_csv('csv/final_datasets/for_xgb/train_test_val/target_test.csv')
y_val = pd.read_csv('csv/final_datasets/for_xgb/train_test_val/target_val.csv')
```

```python
params = {
    'max_depth': 3,
    'learning_rate': 0.013815876417793225,
    'subsample': 0.8564389753793318,
    'colsample_bytree': 0.6720743179899686,
    'reg_lambda': 3.3900178420389695,
    'reg_alpha': 2.2909993262533157,
    'n_estimators': 2000,
    'objective': 'reg:gamma',
    'random_state': 42,
    'early_stopping_rounds': 15,
    'eval_metric': 'mape',
    'enable_categorical': True
}
```

```python
xgb_regressor = xgb.XGBRegressor(**params)
xgb_regressor.fit(
    X_train, y_train,
    eval_set = [(X_val, y_val)],
    verbose = True
)
```

```python
y_train_pred = xgb_regressor.predict(X_train)
y_test_pred = xgb_regressor.predict(X_test)
y_val_pred = xgb_regressor.predict(X_val)
```

```python
response_train = pd.concat([players_train, pd.Series(y_train_pred, name='Value_pred'), y_train], axis=1)
response_test = pd.concat([players_test, pd.Series(y_test_pred, name='Value_pred'), y_test], axis=1)
response_val = pd.concat([players_val, pd.Series(y_val_pred, name='Value_pred'), y_val], axis=1)
```

```python
response_train_val = pd.concat([response_train, response_val])
response = pd.concat([response_train_val, response_test])
```

```python
metrics = pd.DataFrame({
    'MAE':[mae(y_test_pred, y_test), mae(y_train_pred, y_train), mae(y_val_pred, y_val)],
    'RMSE':[rmse(y_test_pred, y_test), rmse(y_train_pred, y_train), rmse(y_val_pred, y_val)],
    'MAPE':[mape(y_test_pred, y_test), mape(y_train_pred, y_train), mape(y_val_pred, y_val)],
    'R2': [r2_score(y_test_pred, y_test), r2_score(y_train_pred, y_train), r2_score(y_val_pred, y_val)]
}, index = ['test', 'train', 'val'])
```

```python
metrics
```
