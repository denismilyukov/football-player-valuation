```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows', None)
```

```python
X_train = pd.read_csv('csv/final_datasets/for_linear_model/FW/predictors_train_filtered.csv')
X_test = pd.read_csv('csv/final_datasets/for_linear_model/FW/predictors_test_filtered.csv')
y_train = pd.read_csv('csv/final_datasets/for_linear_model/FW/target_train_filtered_logariphmed.csv')
y_test = pd.read_csv('csv/final_datasets/for_linear_model/FW/target_test_filtered_logariphmed.csv')
```

```python
train_players = pd.read_csv('csv/final_datasets/for_linear_model/FW/train_players_names.csv')
test_players = pd.read_csv('csv/final_datasets/for_linear_model/FW/test_players_names.csv')
```

```python
all_cols = X_train.columns
all_cols
```

<b>Все переменные</b>

```python
LR_regressor = LinearRegression()
LR_regressor.fit(X_train, y_train)
```

```python
y_pred_log = LR_regressor.predict(X_test)
y_pred = np.exp(y_pred_log)
y_test_original = np.exp(y_test)
```

```python
mae = mean_absolute_error(y_test_original, y_pred)
rmse = np.sqrt(mean_squared_error(y_test_original, y_pred))
r2 = r2_score(y_test_original, y_pred)
mape = mean_absolute_percentage_error(y_test_original, y_pred)
```

```python
coeffs = LR_regressor.coef_[0]
coeffs_df = pd.DataFrame({
    'coeff':coeffs.astype('float')
}, index=X_train.columns)
```

```python
coeffs_df.sort_values(by='coeff', key=abs, ascending=False)
```
