```python
gks = pd.read_csv('../data/raw/stats/combine_gk_24-25.csv')
```

```python
gks
```

```python
duplicates = gks[gks.duplicated(subset=['Player', 'Age', 'Nation'], keep=False)]
duplicates
```

```python
grouped_gk = duplicates.groupby(['Player', 'Age', 'Nation'])
for _, group in grouped_gk:
    #print(group)
    ind_to_save = group.index[0]
    ind_to_del = group.index[1]
    group = group.reset_index(drop=True)
    if group.at[1, '90s'] > group.at[0, '90s']:
        group.at[0, 'Squad'] = group.at[1, 'Squad']
        group.at[0, 'league'] = group.at[1, 'league']
    for col in group.columns[5:len(group.columns) - 2]:
        group.at[0, col] += group.at[1, col]
    group.at[0, 'GA90'] = group.at[0, 'GA'] / group.at[0, '90s']
    group.at[0, 'Save%'] = group.at[0, 'Saves'] / group.at[0, 'SoTA'] * 100
    group.at[0, 'CS%'] = group.at[0, 'CS'] / group.at[0, 'MP'] * 100
    if group.at[0, 'PKatt'] != 0:
        group.at[0, 'PKsv%'] = group.at[0, 'PKsv'] / (group.at[0, 'PKatt'] - group.at[0, 'PKm']) * 100
    gks.loc[ind_to_save] = group.loc[0]
    gks = gks.drop(ind_to_del)
```

```python
gks.to_csv('../data/raw/stats/combine_gk_24-25.csv', index=False)
```
